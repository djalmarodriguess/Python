```python
sobre = pd.read_csv('gender_submission.csv')
base = pd.merge(base, sobre, on='PassengerId')
base
```

```python
df = base 
df
```

```python
# Dividir os dados em conjunto de treinamento e conjunto de teste
x = df.drop('Survived', axis=1)
y = df.Survived
```

# KNN Classifier

```python
# Importar as bibliotecas necessárias
from sklearn.neighbors import KNeighborsClassifier


# Criar o modelo de KNN Classificação
neigh = KNeighborsClassifier(n_neighbors=3)


# Treinar o modelo usando o conjunto de treinamento
neigh.fit(x, y)


# Avaliar a precisão do modelo
accuracy = neigh.score(x,y)


#Imprimir precisão do modelo
print(f"Precisão do classificador KNN: {accuracy * 100:.2f}%")
```

# Tree Classifier

```python
# Importar as bibliotecas necessárias
from sklearn.tree import DecisionTreeClassifier


# Criar o modelo de arvore de classificação
tree = DecisionTreeClassifier(max_depth=2)


# Treinar o modelo usando o conjunto de treinamento
tree.fit(x, y)


# Avaliar a precisão do modelo
accuracy = tree.score(x, y)


#Imprimir precisão do modelo
print(f'A precisão do modelo de árvore de decisão é {accuracy*100:.2f}%' )
```

# Logistic Regression

```python
# Importar as bibliotecas necessárias
from sklearn.linear_model import LogisticRegression


# Criar o modelo de regressão logística
modelo = LogisticRegression()


# Treinar o modelo usando o conjunto de treinamento
modelo.fit(x, y)


# Avaliar a precisão do modelo
precisao = modelo.score(x, y)


#Imprimir precisão do modelo
print(f"A precisão do modelo é:{precisao*100:.2f}%")
```
